## 5. Advanced Functions

```python
def factorial(n):
    return 1 if n == 0 else n * factorial(n-1)

def is_prime(n):
    return n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

def count_vowels(s):
    return sum(1 for char in s.lower() if char in 'aeiou')

def longest_unique_substring(s):
    longest = ""
    for i in range(len(s)):
        sub = ""
        for char in s[i:]:
            if char in sub: break
            sub += char
            longest = max(longest, sub, key=len)
    return longest

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print(f"Vowels in 'Hello World': {count_vowels('Hello World')}")
print(f"Primes in range [10, 20]: {[n for n in range(10, 21) if is_prime(n)]}")
```

## 6. Object-Oriented Programming (OOP)

```python
class Person:
    def __init__(self, name, age):
        self.name, self.age = name, age
    def introduce(self): return f"Hi, I'm {self.name}, {self.age} years old."

class Cat:
    def __init__(self, name, age):
        self.name, self.age = name, age

cats = [Cat("Mestan", 2), Cat("Bella", 5), Cat("Mila", 3)]
oldest = max(cats, key=lambda c: c.age)
print(f"The oldest cat is {oldest.name}, {oldest.age} years old.")
```
