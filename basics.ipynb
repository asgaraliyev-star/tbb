```python
# Base Car class
class Car:
    wheels = 4  # class attribute, bütün avtomobillərdə eyni

    def __init__(self, make, model, year):
        self.make = make        # instance attribute
        self.model = model
        self.year = year

    def describe(self):
        """Return a simple description of the car."""
        return f"This is a {self.year} {self.make} {self.model}"

    def start(self):
        print(f"{self.make} {self.model} is starting...")

    def stop(self):
        print(f"{self.make} {self.model} has stopped.")


# Subclass ElectricCar inherits from Car
class ElectricCar(Car):
    def __init__(self, make, model, year, battery_size, range_km):
        super().__init__(make, model, year)  # call parent constructor
        self.battery_size = battery_size      # in kWh
        self.range_km = range_km              # driving range

    def describe(self):
        """Override parent method to include battery info."""
        base = super().describe()  # get description from Car
        return f"{base} with a {self.battery_size} kWh battery and {self.range_km} km range"

    @property
    def charge_time(self):
        """Estimate charging time based on battery size."""
        # simple estimate: 10 km per kWh per hour of charging
        hours = self.battery_size / 10
        return f"Approx {hours:.1f} hours to fully charge"

    def start(self):
        """Override start to show electric car behavior."""
        print(f"{self.make} {self.model} is now silent but ready to drive!")


# Using the classes
my_car = Car('Ford', 'Mustang', 2020)
print(my_car.describe())
my_car.start()
my_car.stop()

tesla = ElectricCar('Tesla', 'Model Y', 2023, 75, 500)
print(tesla.describe())
print(tesla.charge_time)
tesla.start()
tesla.stop()
```

```python
class BankAccount:
    def __init__(self, owner, initial_balance):
        self.owner = owner
        self.__balance = initial_balance

    def deposit(self, amount):
        self.__balance += amount
        return f'{amount} added. New balance: {self.__balance}'
    
    def withdraw(self, amount):
        if amount > self.__balance:
            return 'Insufficient funds!'
        self.__balance -= amount
        return f'{amount} withdrawn. New balance: {self.__balance}'
    
    def get_balance(self):
        return self.__balance
    
asgar = BankAccount('Asgar',100)
print(asgar.deposit(10))
print(asgar.withdraw(20))
asgar.get_balance()
```

## More Class Practice

```python
import math

class Line:
    def __init__(self, point1, point2):
        self.point1 = point1
        self.point2 = point2

    def calculate_length(self):
        x1, y1 = self.point1
        x2, y2 = self.point2
        return math.sqrt((x2-x1)**2 + (y2-y1)**2)
    
    def calculate_slope(self):
        x1, y1 = self.point1
        x2, y2 = self.point2
        if x1==x2:
            return None
        return (y2-y1)/(x2-x1)
    
l = Line((0, 0), (0, 3))
print("Length:", l.calculate_length()) 
print("Slope:", l.calculate_slope())  
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def calculate_area(self):
        return f'Area: {self.width * self.height}'
    
    def calculate_perimeter(self):
        return f'Perimeter: {2 * (self.width + self.height)}'
    
rect = Rectangle(5, 10)
print(rect.calculate_area())
print(rect.calculate_perimeter())  
```

```python
import math

class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def calculate_volume(self):
        return math.pi * (self.radius**2) * self.height
    
    def calculate_surface_area(self):
        return 2 * math.pi * self.radius * (self.radius + self.height)
    
cyl = Cylinder (3,7)
print(f'Volume: {cyl.calculate_volume():.2f}')
print(f'Surface Area: {cyl.calculate_surface_area():.2f}')
```

## Decorators

```python
def shout(function):
    def wrapper():
        return function().upper()
    return wrapper

@shout
def greet():
    return "Hello, I am a function."

print(greet()) 
```

```python
def repeat(n):
    def decorator(func):
        def wrapper():
            for _ in range(n):
                func()
        return wrapper
    return decorator


@repeat(3)
def say_hi():
    print("Hi!")


say_hi()
```
